```python
def bernoulli_pmf(k, p):
    """
    Args:
    k (int): The outcome(0 or 1)
    p (float): The probability of success (0 <= p <= 1)
    Returns:
    float: The probability of the outcome k 
    """
    if k == 1:
        return p
    elif k == 0:
        return 1 - p
    else:
      return 0 
```

```python
def bernoulli_pmf_concise(x, p):
    return p**x * (1 - p)**(1 - x)
```

```python
Q 5) Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def visualize_type1_type2(mu0, mu1, sigma, n, alpha):
    """
    Visualize Type I and Type II errors for a two-tailed Z-test
    mu0   : Mean under null hypothesis (H0)
    mu1   : Mean under alternative hypothesis (H1)
    sigma : Population standard deviation
    n     : Sample size
    alpha : Significance level
    """
    # Standard error
    se = sigma / np.sqrt(n)
    # Critical values
    z_crit = norm.ppf(1 - alpha / 2)
    lower_crit = mu0 - z_crit * se
    upper_crit = mu0 + z_crit * se

    # Range for plotting
    x = np.linspace(min(mu0, mu1) - 4*se, max(mu0, mu1) + 4*se, 500)

    # Distributions
    y_null = norm.pdf(x, mu0, se)
    y_alt = norm.pdf(x, mu1, se)

    # Type II error ()
    beta = norm.cdf(upper_crit, mu1, se) - norm.cdf(lower_crit, mu1, se)
    # Plot
    plt.figure()
    plt.plot(x, y_null, label="H0: Null Distribution")
    plt.plot(x, y_alt, label="H1: Alternative Distribution")
    plt.axvline(lower_crit, linestyle='--')
    plt.axvline(upper_crit, linestyle='--')

    plt.title("Type I and Type II Errors in Hypothesis Testing")
    plt.xlabel("Sample Mean")
    plt.ylabel("Probability Density")
    plt.legend()
    plt.show()
```

```python
Q 9) Write a Python function to calculate the confidence interval for a sample mean and explain its significance
```

```python
import numpy as np
from scipy.stats import t

def confidence_interval(data, confidence=0.95):
    """
    Calculate confidence interval for a sample mean
    
    data       : list or array of sample values
    confidence : confidence level (default = 0.95)
    """
    data = np.array(data)
    n = len(data)
    mean = np.mean(data)
    std = np.std(data, ddof=1)
    
    t_crit = t.ppf((1 + confidence) / 2, df=n-1)
    
    margin = t_crit * (std / np.sqrt(n))
    
    return mean - margin, mean + margin
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import f_oneway

group1 = np.array([12, 14, 15, 13, 14, 16, 15])
group2 = np.array([18, 19, 17, 20, 21, 19, 18])
group3 = np.array([22, 24, 23, 25, 26, 24, 23])

alpha = 0.05

f_stat, p_value = f_oneway(group1, group2, group3)

print("One-Way ANOVA Results")
print("----------------------")
print(f"F-statistic : {f_stat:.4f}")
print(f"P-value     : {p_value:.4f}")

if p_value < alpha:
    print("Conclusion  : Reject the null hypothesis")
    print("Interpretation: At least one group mean is significantly different.")
else:
    print("Conclusion  : Fail to reject the null hypothesis")
    print("Interpretation: No significant difference among group means.")
plt.figure()
plt.boxplot([group1, group2, group3],
            labels=['Group 1', 'Group 2', 'Group 3'])
plt.title("Boxplot Comparison of Group Means")
plt.xlabel("Groups")
plt.ylabel("Values")
plt.show()
```

```python
Q 15) Simulte random data from a normal distribution, then perform hypothesis testing to evaluate the means
```

```python
import numpy as np
from scipy.stats import ttest_1samp

np.random.seed(0)
data = np.random.normal(loc=50, scale=10, size=30)

mu0 = 48
alpha = 0.05

t_stat, p_value = ttest_1samp(data, mu0)

print("Sample Mean:", round(np.mean(data), 2))
print("Hypothesized Mean:", mu0)
print("T-statistic:", round(t_stat, 4))
print("P-value:", round(p_value, 4))

if p_value < alpha:
    print("Conclusion: Reject the null hypothesis")
    print("Interpretation: Sample mean is significantly different from hypothesized mean.")
else:
    print("Conclusion: Fail to reject the null hypothesis")
    print("Interpretation: No significant difference between sample mean and hypothesized mean.")
```

```python
Q 16) Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results
```

```python
import numpy as np
from scipy.stats import chi2

data = [12, 15, 14, 10, 13, 16, 14, 15, 13]
n = len(data)

sigma0_sq = 4    
alpha = 0.05     

s_sq = np.var(data, ddof=1)

chi_square = (n - 1) * s_sq / sigma0_sq

chi2_lower = chi2.ppf(alpha / 2, df=n - 1)
chi2_upper = chi2.ppf(1 - alpha / 2, df=n - 1)

print("Chi-square Test for Population Variance")
print("---------------------------------------")
print(f"Sample Variance        : {s_sq:.4f}")
print(f"Chi-square Statistic   : {chi_square:.4f}")
print(f"Critical Values        : ({chi2_lower:.4f}, {chi2_upper:.4f})")

if chi_square < chi2_lower or chi_square > chi2_upper:
    print("Conclusion: Reject the null hypothesis")
    print("Interpretation: Population variance is significantly different from the hypothesized variance.")
else:
    print("Conclusion: Fail to reject the null hypothesis")
    print("Interpretation: No significant difference between sample variance and hypothesized variance.")
```

```python
Q 17) Write a Python script to perform a Z-test for comparing proportions between two datasets or groups 
```

```python
from math import sqrt
from scipy.stats import norm

# Data
x1, n1 = 40, 100  
x2, n2 = 30, 100   
alpha = 0.05

p1 = x1 / n1
p2 = x2 / n2
p = (x1 + x2) / (n1 + n2)   

z = (p1 - p2) / sqrt(p * (1 - p) * (1/n1 + 1/n2))

p_value = 2 * (1 - norm.cdf(abs(z)))

print("Z value:", round(z, 3))
print("P-value:", round(p_value, 4))

if p_value < alpha:
    print("Reject H0: proportions are different")
else:
    print("Fail to reject H0: no difference")
```

```python
Q 18) Implement an F-test for comparing the variances of two datasets, then interpret and visualize the results
```

```python
import numpy as np
from scipy.stats import f
x = np.array([12, 14, 15, 13, 14])
y = np.array([18, 20, 22, 19, 21])

alpha = 0.05

# Variances
v1 = np.var(x, ddof=1)
v2 = np.var(y, ddof=1)

# F statistic
F = v1 / v2 if v1 > v2 else v2 / v1

df1 = len(x) - 1
df2 = len(y) - 1

# Critical value
F_crit = f.ppf(1 - alpha, df1, df2)

print("F value:", round(F, 3))
print("Critical value:", round(F_crit, 3))

if F > F_crit:
    print("Reject H0  variances are different")
else:
    print("Fail to reject H0  variances are equal")
```

```python
Q 19) Perform a Chi-square test for goodness of fit with simulated data and analyze the results.
```

```python
import numpy as np
from scipy.stats import chisquare
observed = np.array([8, 9, 10, 11, 12, 10])

expected = np.array([10, 10, 10, 10, 10, 10])

chi_stat, p_value = chisquare(observed, expected)

alpha = 0.05

print("Chi-square statistic:", round(chi_stat, 4))
print("P-value:", round(p_value, 4))

if p_value < alpha:
    print("Conclusion: Reject the null hypothesis")
    print("Interpretation: Data does not fit the expected distribution.")
else:
    print("Conclusion: Fail to reject the null hypothesis")
    print("Interpretation: Data fits the expected distribution.")
```

```python
Q 20) Perform a two-way ANOVA test using Python to study the interaction between two factors and visualize the 
results
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.formula.api import ols

data = {
    "Score": [70, 72, 68, 75, 78, 80, 74, 76,
              65, 67, 66, 69, 72, 74, 71, 73],
    "Method": ["A1"]*8 + ["A2"]*8,
    "StudyTime": ["Low"]*4 + ["High"]*4 + ["Low"]*4 + ["High"]*4
}

df = pd.DataFrame(data)

model = ols('Score ~ C(Method) * C(StudyTime)', data=df).fit()
anova_table = sm.stats.anova_lm(model, typ=2)

print("Two-Way ANOVA Results")
print(anova_table)

means = df.groupby(["Method", "StudyTime"])["Score"].mean().unstack()

plt.figure()
plt.plot(means.index, means["Low"], marker='o', label="Low Study Time")
plt.plot(means.index, means["High"], marker='o', label="High Study Time")
plt.xlabel("Teaching Method")
plt.ylabel("Mean Score")
plt.title("Two-Way ANOVA Interaction Plot")
plt.legend()
plt.show()
```
